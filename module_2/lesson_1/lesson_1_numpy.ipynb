### Mini Project — NumPy Data Analysis

#### Task 1
##### Create a NumPy array representing 10 days of temperatures:

```python
temps = np.array([70, 43, 89, 63, 45, 48, 28, 39, 50, 67])
temps
```

#### Task 2
##### Compute Basic Statistics:

```python
mean_temp = np.mean(temps)
std_temp = np.std(temps)
min_temp = np.min(temps)
max_temp = np.max(temps)
mean_temp, std_temp, max_temp, min_temp
```

### Task 3
#### Using slicing, extract:
#### the first 3 days
#### the last 3 days
#### the middle 4 days (positions 3–6)

```python
first_3_days = temps[0:3]
first_3_days
```

```python
last_3_days = temps[-3:]
last_3_days
```

```python
middle_4_days = temps[3:7]
middle_4_days
```

### Task 4
#### Normalize the temperatures in the dataset

```python
normalized = (temps - np.min(temps)) / (np.max(temps) - np.min(temps))
normalized
```

### Task 5
#### Reshape the data into a 2 row matrix

```python
new_temps = temps.reshape(2,5)
new_temps
```
